La **regresión logística** es uno de los métodos más fundamentales y ampliamente utilizados en Machine Learning. Aunque a menudo se clasifica como un modelo de regresión, es esencialmente un modelo de clasificación binaria, diseñado para predecir probabilidades.

---

### **¿Qué es la Regresión Logística?**

La **regresión logística** es un modelo estadístico y de machine learning ampliamente utilizado para predecir la probabilidad de una clase binaria (por ejemplo, "0" o "1"). Nació en el ámbito de la estadística aplicada a procesos biológicos, pero ha sido generalizada para múltiples áreas, incluidas ciencias sociales, economía y aprendizaje automático.

#### **Orígenes**
- Desarrollada para modelar procesos en biología en el siglo XIX.
- Se basa en la idea de transformar las salidas continuas de una regresión lineal en valores probabilísticos dentro del rango \([0,1]\), usando una función sigmoide.

### **Fundamento Matemático**
La **regresión logística** modela la probabilidad condicional de una clase binaria $ ( P(y=1 | x) )$ , donde:
- $ ( x )$  es un vector de características de entrada.
- $ ( y \in \{0, 1\} )$  es la variable objetivo.

La probabilidad se calcula con la ecuación:

$$
P(y=1|x) = \sigma(w^T x + b)
$$

Donde:
- $ ( \sigma(z) = \frac{1}{1 + e^{-z}} )$  es la **función sigmoide**.
- $ ( w )$ : vector de pesos del modelo (coeficientes de cada característica).
- $ ( b )$ : sesgo o término independiente.
- $ ( z = w^T x + b )$ : combinación lineal de las entradas.


#### **Función Sigmoide**
La **función sigmoide** desempeña un papel crucial al transformar la combinación lineal ($ (z)$ ) en una probabilidad. Su definición es:

$$
\sigma(z) = \frac{1}{1 + e^{-z}}
$$

##### **Propiedades:**
1. **Dominio**: $ ( z \in \mathbb{R} )$  (números reales).
2. **Rango**: $ ( \sigma(z) \in (0,1) )$  (valores estrictamente entre 0 y 1).
3. **Límite en extremos**:
   - Si $ ( z \to +\infty )$ , entonces $ ( \sigma(z) \to 1 )$ .
   - Si $ ( z \to -\infty )$ , entonces $ ( \sigma(z) \to 0 )$ .
4. **Simetría**: Es simétrica alrededor del origen, con $ (\sigma(-z) = 1 - \sigma(z))$ .



#### **Deducción desde la Regresión Lineal**
El modelo surge como una transformación de la **regresión lineal**. Si definimos una combinación lineal $ ( z = w^T x + b )$ , esta puede tomar cualquier valor en $ ( \mathbb{R} )$ , lo cual no es adecuado para representar probabilidades, que deben estar en $ ([0,1])$ . Para ajustarlo:
1. Se aplica la función sigmoide $ ( \sigma(z) )$ , que garantiza que la salida estará en el rango válido.
2. Esto permite interpretar $ ( \sigma(z) )$  como una probabilidad.
